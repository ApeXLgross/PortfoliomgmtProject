## To calculate the reaction of market to the central bank policiy meeting. It does calcaute the daily closing percentage at the meeeting . rolling 7 days clsoing percentage before central bank meeting and post 7 days rolling an save as excel file

```python
import pandas as pd
import numpy as np

# Assuming `european` is your DataFrame with daily log returns for 'euro_bond' and 'euro_aktien'
# Example structure: ['euro_bond', 'euro_aktien'], indexed by datetime

# ECB rate event dates
ecb_rate_hike_dates = pd.to_datetime([
    "2022-07-21", "2022-09-08", "2022-10-27", "2022-12-15",
    "2023-02-02", "2023-03-16", "2023-05-04", "2023-06-15",
    "2023-07-27", "2023-09-14"
])

ecb_rate_cut_dates = pd.to_datetime([
    "2024-06-20", "2024-07-25", "2024-09-12", "2024-10-24", "2024-12-12"
])

ecb_no_change_dates = pd.to_datetime([
    "2023-10-26", "2023-12-14", "2024-02-08", "2024-03-28", "2024-05-09"
])

# Function to calculate returns and rolling averages
def calculate_event_returns(data, event_dates, rolling_window=7):
    results = []

    for event_date in event_dates:
        if event_date in data.index:
            # Get event return
            event_return = data.loc[event_date]

            # Calculate rolling averages before and after the event
            pre_event_mean = data.loc[
                event_date - pd.Timedelta(days=rolling_window):event_date - pd.Timedelta(days=1)
            ].mean()
            post_event_mean = data.loc[
                event_date + pd.Timedelta(days=1):event_date + pd.Timedelta(days=rolling_window)
            ].mean()

            # Combine results into a dictionary for DataFrame conversion
            result = {
                'date': event_date,
                'event_type': None,  # This will be set later
            }
            for asset in data.columns:
                result[f'{asset}_event_return'] = event_return[asset]
                result[f'{asset}_pre_event_mean'] = pre_event_mean[asset]
                result[f'{asset}_post_event_mean'] = post_event_mean[asset]
            results.append(result)

    return pd.DataFrame(results)

# Calculate for each type of event
hike_df = calculate_event_returns(european, ecb_rate_hike_dates)
hike_df['event_type'] = 'rate_hike'

cut_df = calculate_event_returns(european, ecb_rate_cut_dates)
cut_df['event_type'] = 'rate_cut'

pause_df = calculate_event_returns(european, ecb_no_change_dates)
pause_df['event_type'] = 'no_change'

# Combine all results into one DataFrame
all_events_df = pd.concat([hike_df, cut_df, pause_df], ignore_index=True)

# Save the DataFrame to an Excel file
output_file = "ecb_event_analysis.xlsx"
all_events_df.to_excel(output_file, index=False)

print(f"Data has been saved to {output_file}")
```

```python
import pandas as pd
import numpy as np

# Assuming `american` is your DataFrame with daily log returns for 'us_bond' and 'us_stocks'
# Example structure: ['us_bond', 'us_stocks'], indexed by datetime

# Fed rate event dates corrected
fed_rate_hike_dates = pd.to_datetime([
    "2022-03-16",  # 0.25% to 0.25%-0.50%
    "2022-06-15",  # 0.75% to 1.50%-1.75%
    "2022-07-27",  # 0.75% to 2.25%-2.50%
    "2022-09-21",  # 0.75% to 3.00%-3.25%
    "2022-11-02",  # 0.75% to 3.75%-4.00%
    "2023-02-01",  # 0.25% to 4.50%-4.75%
    "2023-03-22",  # 0.25% to 4.75%-5.00%
    "2023-05-03",  # 0.25% to 5.00%-5.25%
    "2023-07-26"   # 0.25% to 5.25%-5.50%
])

fed_no_change_dates = pd.to_datetime([
    "2023-09-20",  # Held at 5.25%-5.50%
    "2023-11-01",  # Held at 5.25%-5.50%
    "2024-01-31",  # Held at 5.25%-5.50%
    "2024-03-20",  # Held at 5.25%-5.50%
    "2024-05-01",  # Held at 5.25%-5.50%
    "2024-06-19",  # Held at 5.25%-5.50%
    "2024-07-31"   # Held at 5.25%-5.50%
])

fed_rate_cut_dates = pd.to_datetime([
    "2024-09-18",  # -0.25% to 5.00%-5.25%
    "2024-11-06",  # -0.25% to 4.75%-5.00%
    "2024-12-11"   # -0.25% to 4.50%-4.75%
])

# Function to calculate returns and rolling averages
def calculate_event_returns(data, event_dates, rolling_window=7):
    results = []

    for event_date in event_dates:
        if event_date in data.index:
            # Get event return
            event_return = data.loc[event_date]

            # Calculate rolling averages before and after the event
            pre_event_mean = data.loc[
                event_date - pd.Timedelta(days=rolling_window):event_date - pd.Timedelta(days=1)
            ].mean()
            post_event_mean = data.loc[
                event_date + pd.Timedelta(days=1):event_date + pd.Timedelta(days=rolling_window)
            ].mean()

            # Combine results into a dictionary for DataFrame conversion
            result = {
                'date': event_date,
                'event_type': None,  # This will be set later
            }
            for asset in data.columns:
                result[f'{asset}_event_return'] = event_return[asset]
                result[f'{asset}_pre_event_mean'] = pre_event_mean[asset]
                result[f'{asset}_post_event_mean'] = post_event_mean[asset]
            results.append(result)

    return pd.DataFrame(results)

# Calculate for each type of event
hike_df = calculate_event_returns(american, fed_rate_hike_dates)
hike_df['event_type'] = 'rate_hike'

cut_df = calculate_event_returns(american, fed_rate_cut_dates)
cut_df['event_type'] = 'rate_cut'

pause_df = calculate_event_returns(american, fed_no_change_dates)
pause_df['event_type'] = 'no_change'

# Combine all results into one DataFrame
all_events_df = pd.concat([hike_df, cut_df, pause_df], ignore_index=True)

# Add rate change information with corrected values
rate_changes = {
    '2022-03-16': 0.25,  # First hike
    '2022-06-15': 0.75,  # Second hike
    '2022-07-27': 0.75,  # Third hike
    '2022-09-21': 0.75,  # Fourth hike
    '2022-11-02': 0.75,  # Fifth hike
    '2023-02-01': 0.25,  # Sixth hike
    '2023-03-22': 0.25,  # Seventh hike
    '2023-05-03': 0.25,  # Eighth hike
    '2023-07-26': 0.25,  # Ninth hike
    '2024-09-18': -0.25, # First cut
    '2024-11-06': -0.25, # Second cut
    '2024-12-11': -0.25  # Third cut
}

# Add rate change column to DataFrame
all_events_df['rate_change'] = all_events_df['date'].dt.strftime('%Y-%m-%d').map(rate_changes)

# Add final rate range column
rate_ranges = {
    '2022-03-16': '0.25%-0.50%',
    '2022-06-15': '1.50%-1.75%',
    '2022-07-27': '2.25%-2.50%',
    '2022-09-21': '3.00%-3.25%',
    '2022-11-02': '3.75%-4.00%',
    '2023-02-01': '4.50%-4.75%',
    '2023-03-22': '4.75%-5.00%',
    '2023-05-03': '5.00%-5.25%',
    '2023-07-26': '5.25%-5.50%',
    '2024-09-18': '5.00%-5.25%',
    '2024-11-06': '4.75%-5.00%',
    '2024-12-11': '4.50%-4.75%'
}
all_events_df['rate_range'] = all_events_df['date'].dt.strftime('%Y-%m-%d').map(rate_ranges)

# Sort by date
all_events_df = all_events_df.sort_values('date')

# Save the DataFrame to an Excel file
output_file = "fed_event_analysis.xlsx"
all_events_df.to_excel(output_file, index=False)

print(f"Data has been saved to {output_file}")

```
